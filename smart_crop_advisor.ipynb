```python
import pandas as pd

data = pd.read_csv("Crop_recommendation.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = data.drop("label", axis=1)
y = data["label"]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train,y_train)

print("Model trained successfully")
```

```python
input_data = pd.DataFrame({
"N":[90],
"P":[40],
"K":[40],
"temperature":[25],
"humidity":[85],
"ph":[6.5],
"rainfall":[200]
})#here we are giving input data,from this data the later calculation are done.
```

```python
prediction = model.predict(input_data)
```

```python
def disease_warning(crop, temperature, humidity, rainfall):

    if crop == "rice":
        if humidity > 80 and rainfall > 150:
            return "High risk of Rice Blast (fungal infection). Use fungicide and monitor leaves."

    if crop == "potato":
        if humidity > 85 and temperature < 25:
            return "Risk of Late Blight fungus. Ensure proper drainage."

    if crop == "wheat":
        if humidity > 70:
            return "Possible Wheat Rust infection."

    return "No major infection risk detected."
```

```python
crop = prediction[0]

warning = disease_warning(crop,25,85,200)

print("Recommended crop:", crop)
print("Disease warning:", warning)
```
